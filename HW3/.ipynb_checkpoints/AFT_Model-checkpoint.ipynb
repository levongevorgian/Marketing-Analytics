## Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from lifelines import WeibullAFTFitter, LogLogisticAFTFitter, LogNormalAFTFitter
```

## Data & Preprocessing

```python
df = pd.read_csv("telco.csv")

df = df.dropna(subset=["tenure", "churn"])
df = df[df["tenure"] > 0].copy()

df["churn"] = df["churn"].map({"Yes": 1, "No": 0})

df["churn"] = pd.to_numeric(df["churn"], errors="coerce")
df = df.dropna(subset=["churn"])

if "ID" in df.columns:
    df_ids = df["ID"].copy()
    df = df.drop(columns=["ID"])
else:
    df_ids = pd.Series(np.arange(len(df)), name="ID")

obj_cols = df.select_dtypes(include="object").columns.tolist()

obj_cols = [c for c in obj_cols if c not in ["churn", "tenure"]]
df_encoded = pd.get_dummies(df, columns=obj_cols, drop_first=True)

duration_col = "tenure"
event_col = "churn"
```

## Fitting AFT Models

```python
models = {
    "WeibullAFT": WeibullAFTFitter(),
    "LogLogisticAFT": LogLogisticAFTFitter(),
    "LogNormalAFT": LogNormalAFTFitter()
}

aic_scores = {}
fitted_models = {}

for name, model in models.items():
    model.fit(df_encoded, duration_col=duration_col, event_col=event_col)
    aic_scores[name] = model.AIC_
    fitted_models[name] = model
    print(f"\n{name} Summary:")
    display(model.summary)  

print("\nAIC Scores:")
for name, aic in aic_scores.items():
    print(f"{name}: {aic:.4f}")
```

## Survival Curves

```python
feature_columns = [c for c in df_encoded.columns if c not in [duration_col, event_col]]
median_customer = df_encoded[feature_columns].median().to_frame().T

plt.figure(figsize=(10,6))
for name, model in fitted_models.items():
    surv = model.predict_survival_function(median_customer)
    plt.plot(surv.index, surv.iloc[:,0], label=name)

plt.title("AFT Survival Curves (Median Customer)")
plt.xlabel("Tenure (Months)")
plt.ylabel("Survival Probability")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Selecting AFT Model

```python
best_model_name = min(aic_scores, key=aic_scores.get)
final_model = fitted_models[best_model_name]
print(f"\nFinal Model Selected: {best_model_name}")

summary = final_model.summary.copy()

sig_rows = summary[(summary.index.get_level_values('covariate') != 'Intercept') & (summary['p'] < 0.05)]
sig_covariates = sig_rows.index.get_level_values('covariate').unique().tolist()
print("Significant covariates (p < 0.05):", sig_covariates)

sig_covariates = [c for c in sig_covariates if c not in [duration_col, event_col]]

if len(sig_covariates) == 0:
    print("No significant covariates found at p<0.05; keeping full model.")
else:
    final_model.fit(df_encoded[sig_covariates + [duration_col, event_col]],
                    duration_col=duration_col, event_col=event_col)
    print("Refitted final model with significant covariates.")

model_features = sig_covariates
```

## Survival Analysis by Region

```python
plt.figure(figsize=(10,6))

for reg in df["region"].unique():
    mask = df["region"] == reg
    median_reg = df_encoded.loc[mask, model_features].median().to_frame().T
    surv_reg = final_model.predict_survival_function(median_reg)
    plt.plot(surv_reg.index, surv_reg.iloc[:,0], label=f"Region {reg}")

plt.title("Survival Curves by Region (Final Model)")
plt.xlabel("Tenure (Months)")
plt.ylabel("Survival Probability")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Predicted Lifetime and CLV

```python
predicted_lifetime = final_model.predict_median(df_encoded[model_features])
df_encoded["predicted_lifetime"] = predicted_lifetime.values

monthly_revenue = 50.0    
discount_rate = 0.01       

df_encoded["predicted_lifetime_capped"] = df_encoded["predicted_lifetime"].clip(lower=0.1, upper=120)

t = df_encoded["predicted_lifetime_capped"]
df_encoded["CLV"] = (monthly_revenue * (1 - (1 + discount_rate) ** (-t))) / discount_rate
```

## CLV Distribution 

```python
plt.figure(figsize=(10,5))
plt.hist(df_encoded["CLV"], bins=40)
plt.title("CLV Distribution")
plt.xlabel("CLV")
plt.ylabel("Count")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,4))
plt.boxplot(df_encoded["CLV"])
plt.title("CLV Boxplot")
plt.show()
```

## CLV by Segments

```python
if "region" in df.columns:
    clv_by_region = df_encoded.join(df["region"]).groupby("region")["CLV"].mean().sort_values(ascending=False)
    print("\nCLV by Region:")
    print(clv_by_region)
else:
    region_cols = [c for c in df_encoded.columns if c.startswith("region_")]
    if region_cols:
        for rc in region_cols:
            print(f"Avg CLV for {rc}: {df_encoded.loc[df_encoded[rc]==1, 'CLV'].mean()}")
```

## CLV by Region

```python
clv_by_region.plot(kind="bar", figsize=(8,5))
plt.title("Average CLV by Region")
plt.ylabel("Average CLV")
plt.grid(True)
plt.show()

if "custcat" in df.columns:
    clv_by_custcat = df_encoded.join(df["custcat"]).groupby("custcat")["CLV"].mean()
    clv_by_custcat.plot(kind="bar", figsize=(8,5))
    plt.title("Average CLV by Customer Category")
    plt.ylabel("Average CLV")
    plt.grid(True)
    plt.show()
```

## Estimated Retention Budget

```python
surv_12 = final_model.predict_survival_function(df_encoded[model_features], times=[12])
surv_at_12 = surv_12.loc[12].values  
churn_prob_12 = 1 - surv_at_12

expected_budget = (churn_prob_12 * df_encoded["CLV"].values).sum()
print(f"\nSuggested Annual Retention Budget: ${expected_budget:.2f}")
```

## Churn vs CLV

```python
plt.figure(figsize=(8,6))
plt.scatter(df_encoded["CLV"], churn_prob_12)
plt.title("Churn Probability (12m) vs CLV")
plt.xlabel("CLV")
plt.ylabel("Churn Probability (12m)")
plt.grid(True)
plt.show()
```

## Final Dataframe

```python
out = df_encoded.copy()
out["ID"] = df_ids.values
out.to_csv("telco_encoded_with_clv.csv", index=False)
```
